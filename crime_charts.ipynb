```python
import pandas as pd
import numpy as np
import csv
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
crime_df = pd.read_csv('total_crime_commarea.csv')
```

```python
crime_df.dtypes
```

```python
#crime_sides_pop['Pop_2010'] = crime_sides_pop['Pop_2010'].apply(pd.to_numeric)
```

```python
plt.scatter(crime_df['Pop_2010'], crime_df['Crimes_Reported'], marker = "o", facecolors="red", edgecolors="black")
plt.title("Crime per Population - Neighborhood")
plt.xlabel("Population")
plt.ylabel("Crime")
plt.show()
plt.savefig('crime_scatter.png')
```

```python
bar_chart = crime_df.groupby('Side').Crimes_Reported.sum()
```

```python
bar_chart = bar_chart.reset_index()
```

```python
side = bar_chart["Side"]
```

```python
x_axis = np.arange(len(side))
```

```python
bar_chart_pop = crime_df.groupby('Side').Pop_2010.sum()
```

```python
bar_chart_pop = bar_chart_pop.reset_index()
```

```python
side_pop = bar_chart_pop["Side"]
```

```python
plt.bar(x_axis, bar_chart_pop['Pop_2010'], color = "blue", label = "Population")
plt.bar(x_axis, bar_chart['Crimes_Reported'], color = "red", label = "Crimes")
plt.xticks(x_axis,side_pop, rotation=15)
plt.legend(loc="best")
plt.title("Crimes per Population - Sides")
plt.xlabel("Chicago - Sides")
plt.ylabel("Population & Crimes")
plt.show()
plt.savefig('crime_bar.png')
```
