```python
import pandas as pd
import numpy as np
import csv
```

```python
data_file = pd.read_csv('crimes_-_2001_to_present.csv')
```

```python
data_file.head()
```

```python
data_reduce = data_file[['ID', 'Community Area', 'Year']]
```

```python
data_reduce = data_reduce.loc[data_reduce.Year > 2009]
```

```python
data_reduce = data_reduce.loc[data_reduce.Year < 2011]
```

```python
data_reduce = data_reduce.rename(columns = {'ID': 'Crimes_Reported', 'Community Area' : 'Area'})
```

```python
data_reduce = data_reduce['Crimes_Reported'].groupby([data_reduce['Area']]).count()
```

```python
comm_crime_data = pd.DataFrame(data_reduce)
```

```python
#comm_crime_data.to_csv('total_crime_commarea.csv')
```

```python
comm_crime_data.head(10)
```

```python
pop_data = pd.read_csv('census_data.csv')
```

```python
pop_data.head(10)
```

```python
pop_data_reduce = pop_data[['GEOG', '2010_POP']]
```

```python
pop_data_reduce = pop_data_reduce.rename(columns = {'GEOG': 'Name', '2010_POP': 'Pop_2010'})
```

```python
comm_sides = pd.read_csv('chicago_community_sides.csv')
```

```python
comm_sides.head(10)
```

```python
comm_pop_df = pd.merge(pop_data_reduce, comm_sides, how = 'left', on = 'Name')
```

```python
comm_pop_df.head()
```

```python
comm_pop_crime = pd.merge(comm_pop_df, comm_crime_data, how = 'left', on = 'Area')
```

```python
comm_pop_crime.head(10)
```

```python
comm_pop_crime.to_csv('total_crime_commarea.csv')
```
